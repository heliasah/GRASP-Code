```python
# Import necessary libraries
import pandas as pd
from sklearn.model_selection import train_test_split
from google.colab import files

# Step 1: Upload your file to Colab
uploaded = files.upload()  # This will prompt you to upload the file

# Step 2: Read the CSV file
file_name = list(uploaded.keys())[0]  # Get the uploaded file name
df = pd.read_csv(file_name, sep=';', header=None)  # Specify separator and no header

# Add column names for better readability
df.columns = ['ID', 'Gender', 'Value1', 'Value2', 'Education', 'Score', 'Height', 'BMI']

# Step 3: Split the dataset into 80% Train and 20% Test
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Step 4: Separate males and females in both train and test sets
train_males = train_df[train_df['Gender'] == 'male']
train_females = train_df[train_df['Gender'] == 'female']
test_males = test_df[test_df['Gender'] == 'male']
test_females = test_df[test_df['Gender'] == 'female']

# Step 5: Save the files
train_males.to_csv('train_males.csv', index=False)
train_females.to_csv('train_females.csv', index=False)
test_males.to_csv('test_males.csv', index=False)
test_females.to_csv('test_females.csv', index=False)

# Step 6: Download the files to your system
files.download('train_males.csv')
files.download('train_females.csv')
files.download('test_males.csv')
files.download('test_females.csv')
```

```python
import os
import numpy as np
import shutil
from google.colab import files

# Step 1: Define input and output folders
male_matrices_folder = 'males_test_matrices'  # Folder containing CSV files
output_folder = 'converted_males_test_matrices'  # Folder to save .npy files
zip_file_name = 'converted_males_test_matrices.zip'  # ZIP file name

# Create the output folder if it doesn't exist
os.makedirs(output_folder, exist_ok=True)

# Step 2: Get the list of all CSV files (connectivity matrices)
male_matrix_files = [os.path.join(male_matrices_folder, file)
                       for file in os.listdir(male_matrices_folder) if file.endswith('.csv')]

# Step 3: Convert each CSV file to .npy format and save in the new folder
for matrix_file in male_matrix_files:
    # Load the CSV file as a numpy array
    matrix = np.loadtxt(matrix_file, delimiter=',')

    # Create the .npy file name inside the output folder
    npy_file = os.path.join(output_folder, os.path.basename(matrix_file).replace('.csv', '.npy'))

    # Save the matrix as a .npy file
    np.save(npy_file, matrix)
    print(f"Converted {matrix_file} to {npy_file}")

# Step 4: Zip the output folder
shutil.make_archive(output_folder, 'zip', output_folder)

# Step 5: Download the ZIP file
files.download(zip_file_name)
```

```python
import os
import numpy as np
from random import choice
from zipfile import ZipFile

# Step 1: Define the folder containing the .npy files
female_matrices_folder = 'females_train_matrices'  # Update with the correct path to your folder

# Get the list of all .npy files (connectivity matrices)
female_matrix_files = [os.path.join(female_matrices_folder, file)
                       for file in os.listdir(female_matrices_folder) if file.endswith('.npy')]

# Step 2: Generate 1000 A matrices and save them in a zip file
output_zip_file = 'generated_female_train_matrices.zip'

# Create a ZipFile object to store the matrices
with ZipFile(output_zip_file, 'w') as zipf:
    for i in range(1000):
        # Step 3: Create an empty matrix A
        sample_matrix = np.load(female_matrix_files[0])  # Load one matrix to get its dimensions
        matrix_shape = sample_matrix.shape
        A = np.zeros(matrix_shape)

        # Step 4: Fill each element (i, j) of the matrix A
        rows, cols = matrix_shape
        for row in range(rows):
            for col in range(cols):
                # Randomly select one connectivity matrix file
                random_matrix_file = choice(female_matrix_files)

                # Load the selected matrix
                random_matrix = np.load(random_matrix_file)

                # Assign the (row, col) element to matrix A
                A[row, col] = random_matrix[row, col]

        # Step 5: Save the filled matrix A to the zip file
        output_file_A = f'filled_matrix_A_{i+1}.npy'  # Name for each matrix
        np.save(output_file_A, A)

        # Add the generated matrix to the zip file
        zipf.write(output_file_A)

        # Optionally, delete the file after adding it to the zip (cleanup)
        os.remove(output_file_A)

print(f"1000 matrices generated and saved in {output_zip_file}")
```
